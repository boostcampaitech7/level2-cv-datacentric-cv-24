```python
import os
from PIL import Image
import matplotlib.pyplot as plt
from IPython.display import display, clear_output


folder_paths = ["./predictions/best_vis", "./predictions/latest_vis", "./predictions/final_vis"]

image_names = os.listdir(folder_paths[0])

try:
    current_index
except NameError:
    current_index = 0

def plot_next_image_set():
    global current_index
    
    if current_index >= len(image_names):
        print("No more images to display.")
        return
    
    img_name = image_names[current_index]
    images = []
    for folder in folder_paths:
        img_path = os.path.join(folder, img_name)
        image = Image.open(img_path)
        images.append(image)
    
    fig, axes = plt.subplots(1, len(images), figsize=(15, 5))
    fig.suptitle(img_name)
    
    for i, img in enumerate(images):
        axes[i].imshow(img)
        axes[i].axis('off')
        axes[i].set_title(f'Folder {i+1}')
    
    plt.show()
    
    current_index += 1

plot_next_image_set()
```
